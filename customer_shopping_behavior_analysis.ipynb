```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# check null values.
df.isnull().sum()
```

```python
# fill missing or null values.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna( x.median()))
```

```python
df.isnull().sum()
```

```python
# replace column name 
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.info()
```

```python
# create a column age_group.
labels=['Young Adult','Adult','Middle-age','Senior']
df['age_group']=pd.qcut(df['age'],q=4 ,labels=labels)
```

```python
df[['age','age_group']].head(12)
```

```python
#create a column purchase_frequency
frequency_mapping={
    
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 months': 90,
    }
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df['purchase_frequency_days'].dtypes
df['purchase_frequency_days'] = pd.to_numeric(df['purchase_frequency_days'], errors='coerce').fillna(0).astype(int)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['promo_code_used']== df['discount_applied']).all()
```
